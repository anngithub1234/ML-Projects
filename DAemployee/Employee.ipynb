```python
# Target Variable(Attrition)
data['Attrition'] = data['Attrition'].replace({'No':0,'Yes':1})
```

```python
#encode binary variables
data['OverTime'] = data['OverTime'].map({'No':0,'Yes':1})
data['Gender'] = data['Gender'].map({'Male':0,'Female':1})
```

```python
# encode categorical columns which are ordinal, use labelEncoding
# apply Label encoder to df_categorical
from sklearn.preprocessing import LabelEncoder
encoding_cols=['BusinessTravel','Department','EducationField','JobRole','MaritalStatus']
label_encoders = {}
for column in encoding_cols:
    label_encoders[column] = LabelEncoder()
    data[column] = label_encoders[column].fit_transform(data[column])
```

```python
# look at the final data
data.head()
```

```python
X = data.drop(['Attrition','Over18'], axis=1)
y = data['Attrition'].values
```

```python
from collections import Counter
from imblearn.over_sampling import RandomOverSampler
print(Counter(y))
rus = RandomOverSampler(random_state = 42)
X_over, y_over = rus.fit_resample(X,y)
print(Counter(y_over))
```

```python
# Split the data into training and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_over, y_over, test_size=0.2, random_state=42)
```

```python
# Sanity check for the sampled data
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, roc_curve, roc_auc_score
```

```python
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
prediction=logreg.predict(X_test)
cnf_matrix = confusion_matrix(y_test,prediction)
print("Accuracy Score -", accuracy_score(y_test , prediction))
```

```python
fig = plt.figure(figsize = (15,6))
ax1 = fig.add_subplot(1,2,1)
ax1 = sns.heatmap(pd.DataFrame(cnf_matrix), annot = True, cmap = 'Blues', fmt = 'd')
bottom, top = ax1.get_ylim()
ax1.set_ylim(bottom + 0.5, top - 0.5)
plt.xlabel('Predicted')
plt.ylabel('Expected')

ax2 = fig.add_subplot(1,2,2)
y_pred_proba = logreg.predict_proba(X_test)[::,1]
fpr, tpr, _ = roc_curve(y_test,  prediction)
auc = roc_auc_score(y_test, prediction)
ax2 = plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```

```python
# Extract the feature importance (coefficients) from the model
feature_importance = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': logreg.coef_[0]
})

# Sort the features by their importance
feature_importance = feature_importance.sort_values(by='Importance', ascending=False)

# Plot the feature importance
plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data=feature_importance, palette='viridis')
plt.title('Feature Importance from Logistic Regression')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()
```

```python
def predict_employee_attrition(model, label_encoders, employee_input):
    # Transforming the input data
    employee_df = pd.DataFrame([employee_input])

    # Encoding categorical columns
    for column in encoding_cols:
        if column in employee_df.columns and column in label_encoders:
            employee_df[column] = label_encoders[column].transform(employee_df[column])

    # Mapping binary columns
    binary_columns = {
        'OverTime': {'No': 0, 'Yes': 1},
        'Gender': {'Male': 0, 'Female': 1}
    }

    for column, mapping in binary_columns.items():
        if column in employee_df.columns:
            employee_df[column] = employee_df[column].map(mapping)

    # Ensure the order of columns matches the training data
    employee_df = employee_df[X.columns]

    # Making a prediction
    prediction = model.predict(employee_df)
    prediction_proba = model.predict_proba(employee_df)[:, 1]

    # Output result
    return prediction[0], prediction_proba[0]

# Example usage
employee_input = {
    'Age': 28,
    'BusinessTravel': 'Travel_Rarely',
    'DailyRate': 100000,
    'Department': 'Sales',
    'DistanceFromHome': 10,
    'Education': 3,  # Integer value representing 'Bachelor'
    'EducationField': 'Life Sciences',
    'EmployeeCount': 1,
    'EmployeeNumber': 9999,
    'EnvironmentSatisfaction': 3,
    'Gender': 'Female',
    'JobInvolvement': 3,
    'JobLevel': 2,
    'JobRole': 'Sales Executive',
    'JobSatisfaction': 4,
    'MaritalStatus': 'Single',
    'MonthlyIncome': 50000,
    'MonthlyRate': 1500000,
    'NumCompaniesWorked': 1,
    'OverTime': 'No',
    'PercentSalaryHike': 12,
    'PerformanceRating': 3,
    'RelationshipSatisfaction': 3,
    'StockOptionLevel': 1,
    'TotalWorkingYears': 6,
    'TrainingTimesLastYear': 3,
    'WorkLifeBalance': 3,
    'YearsAtCompany': 2,
    'YearsInCurrentRole': 1,
    'YearsSinceLastPromotion': 1,
    'YearsWithCurrManager': 1,
    'HourlyRate': 65,
    'StandardHours': 80
}

prediction, prediction_proba = predict_employee_attrition(logreg, label_encoders, employee_input)
print(f"Prediction: {'Leave' if prediction == 1 else 'Stay'}")
print(f"Probability of leaving: {prediction_proba:.2f}")
```
